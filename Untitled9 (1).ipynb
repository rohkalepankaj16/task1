```python
import pandas as pd

df = pd.read_excel("School data.xlsx")

```

```python
df.head()
```

```python
#Data Understanding
# rows & columns
df.shape 
# data types
df.info() 
# statistical summary
df.describe() 
```

```python
# Data Cleaning
#Check missing values
df.isnull().sum()
```

```python
#Fill missing values (numerical columns with mean)
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Fill categorical missing values
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
#Removing duplicates
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
#EDA
import matplotlib.pyplot as plt
import seaborn as sns

#Univariate analysis

sns.histplot(df["%_Math_Score"], kde=True)
plt.title("Distribution of Math Score")
plt.show()
```

```python
#Bivariate analysis
#Scatter plot
sns.scatterplot(x="%_Math_Score", y="%_Science_Score", data=df)
plt.title("Math vs Science Score")
plt.show()

sns.boxplot(x="Urban_Rural", y="%_Math_Score", data=df)
plt.title("Urban vs Rural vs Math Score")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Insights
#Students with higher attendance tend to score better academically.
#Math and Science scores show strong positive correlation.
#Female students perform slightly better in language subjects.
#Some outliers exist in exam scores indicating exceptional or poor performance.
#Students with low attendance are more likely to have lower overall grades.
```

```python
#Feature Engineering
df["Total_Score"] = df["%_Math_Score"] + df["%_Science_Score"] + df["%_Language_Score"]

df["Average_Score"] = df["Total_Score"] / 3
df.head()
```
